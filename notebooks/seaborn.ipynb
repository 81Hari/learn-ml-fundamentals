## Built-In DataSets

- Built-In datasets of seaborn

### Tip vs Total Bill (Relation-Plot)

```python
tip_df = sns.load_dataset('tips')
tip_df.head()
```

#### Relation Plot

```python
sns.relplot(data=tip_df,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

- seaborn automatically differenticates smoker, non-smoker
- but in matplot we need to do it manually 

#### Setting a Default Theme for Plot
- To set Girds and make the plot more spacious

```python
sns.set_theme()
```

```python
sns.relplot(data=tip_df,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

#### HeatMap (Co-Relation)

- It tells which column is important for prediction

<code>heatmap()</code>

- cbar : colourbar
- square:
- fmt: how much floting point needed
- annot: Annotation
- annot_kws: Annotation size
- cmap: colour of the map




```python
correlation = boston_house_df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
