# MatPlotLib
- useful for making plots

## Importing matplotlib

```python
import matplotlib.pyplot as plt
```

## Importing NumPy
- To get data for our plots

```python
import numpy as np
```

## Creating Data for Ploting

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

## Plotting the Data


### Sin Graph

```python
plt.plot(x,y)
plt.show()
```

### Cos Graph

```python
plt.plot(x,z)
plt.show()
```

### Sin Graph with x-axis and y-axis label

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('Sin Graph')
plt.show()
```

### Parabola

```python
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

### Plotting with symbols

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

### Multiple Graph in single plot

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

### Bar Graph
- used to visuvalize number of category and magitude of that catagory

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','Fench','Spanish','Latin','German']
people = [500,100,200,300,150]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number of People')
plt.show()
```

### Pie Chart
- used to find the distribution

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','Fench','Spanish','Latin','German']
people = [500,100,200,300,150]
ax.pie(people, labels = languages, autopct='%1.1f%%' )
plt.show()
```

### Scatter Plot
- Useful in Clustering

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()

```

### 3-D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
